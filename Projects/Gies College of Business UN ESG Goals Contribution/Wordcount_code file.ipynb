```python
import pandas as pd
import re
```

```python
dictionary = pd.read_excel("Dictionary.xlsx")
df = pd.read_csv("Gies_article_final.csv")
df.head(5)
```

```python
#Dictionary into one list
dict_word = dictionary["Words"].dropna()
dict_list = []
for i in range(len(dict_word)):
    dict_list.append(dict_word[i])
    
for i in range(len(dictionary)):
    dict_list.append(dictionary["Phrases"][i])
    
dict_list
```

```python
# count the number of articles that contain a word/phrase in the dictionary by professor name and dept
count = (
    df.groupby(['name', 'department'])
    .apply(lambda group: group['article'].str.contains('|'.join(dict_list)).sum())
    .reset_index(name='count')
)
count = count[count['count'] != 0]
print(count)
```

```python
count.groupby("department").sum(numeric_only=True)
```

```python
# count the number of articles that match each phrase
count1 = []
for word in dict_list:
    count = df['article'].str.contains(word).sum()
    count1.append({'word': word, 'count': count})

wordcount_df = pd.DataFrame(count1)
wordcount_df = wordcount_df[wordcount_df['count'] != 0]
print(wordcount_df)
```

```python
# count the number of articles for each year
count_year = (
    df.groupby(['year'])
    .apply(lambda group: group['article'].str.contains('|'.join(dict_list)).sum())
    .reset_index(name='count')
)
count_year = count_year[count_year['count'] != 0]
print(count_year)
```
